```python
import pandas as pd

df = pd.read_csv('three_cols.csv', index_col=1, parse_dates=True)
df2 = pd.pivot_table(df, values='Time_work', columns=['Tasks_number'], index="Proc_number")
df2.to_csv('EB_CUT.csv')
```

```python
df2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5, 5), dpi = 111)
sns.heatmap(
    df2.sort_values(by=["Proc_number"], ascending=False), 
    cmap = sns.cm.rocket_r,
    xticklabels = range(100, 2500, 100)
)
plt.ylabel("Proc_number")
plt.xlabel("Tasks_number")

plt.savefig("hetmuuup.jpg", bbox_inches="tight")

plt.show()
```
